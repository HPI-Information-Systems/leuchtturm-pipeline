```python
import pickle
import os
```

```python
PATH_TO_FILES = '../models'
PATH_TO_TM = PATH_TO_FILES + '/topicmodel_dataset.pickle'
PATH_TO_DICT = PATH_TO_FILES + '/topicdict_dataset.pickle'
PATH_TO_CORPUS = PATH_TO_FILES + '/topiccorpus_dataset.pickle'
```

```python
with open(os.path.abspath(PATH_TO_TM), mode='rb') as pfile:
    model = pickle.load(pfile)
with open(os.path.abspath(PATH_TO_DICT), mode='rb') as pfile:
    dictionary = pickle.load(pfile)
with open(os.path.abspath(PATH_TO_CORPUS), mode='rb') as pfile:
    corpus = pickle.load(pfile)
```

```python
import pyLDAvis
import pyLDAvis.gensim
pyLDAvis.enable_notebook()
```

```python
pyLDAvis.gensim.prepare(model, corpus, dictionary)
```
