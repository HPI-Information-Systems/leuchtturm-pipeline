```python
import pickle
import os
import pyLDAvis
import pyLDAvis.gensim
```

```python
PATH_TO_TM = '/topicmodel_enron.pickle'
PATH_TO_DICT = '/topicdict_enron.pickle'
PATH_TO_CORPUS = '/topiccorpus_enron.pickle'

def load_mcd(path_to_files):
    with open(os.path.abspath(path_to_files + PATH_TO_TM), mode='rb') as pfile:
        model = pickle.load(pfile)
    with open(os.path.abspath(path_to_files + PATH_TO_CORPUS), mode='rb') as pfile:
        corpus = pickle.load(pfile)
    with open(os.path.abspath(path_to_files + PATH_TO_DICT), mode='rb') as pfile:
        dictionary = pickle.load(pfile)
    return model, corpus, dictionary
```

```python
path_to_files = '../../historic_tm_dont_delete/5-d5ca4d793af8cb751f29a3f4ad634f04e463f100'
old_model = pyLDAvis.gensim.prepare(*load_mcd(path_to_files))
pyLDAvis.display(old_model)
```
